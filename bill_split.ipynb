```python
#Read file
file = open('values.dat','r')
line = []
for i in file:
    line.append(i.split(' ')[:-1]) #don't include '\n'

#Make dictionary for all people and their owed amount
All = {} 
for i in range(len(line[0])-3):
    All[line[0][i+3]]=0 #Person name are keys and their owed amount is the value
print(All)

#update paid values
for i in range(1,len(line)):
    payer = line[i][1] #payer name
    amt = line[i][2] #paid amount
    if payer in All.keys():
        All[payer] = float(amt)
    else:
        raise ValueError(f'Payer \'{payer}\' not found list of people, please check spelling (case sensitive).')
print(All)

#update owed values
for j in range(1,len(line)):
    if 'e' in line[j]:
        num_e = line[j].count('e') #number of time 'e' occurs
        eq_amt = float(line[j][2])/num_e #the equal split amount
        for i in range(3,len(line[0])):
            pers = line[0][i]
            if line[j][i] == 'e':
                All[pers] -= eq_amt
            else:
                All[pers] -= float(line[j][i])
    else:
        for i in range(3,len(line[0])):
            pers = line[0][i]
            All[pers] -= float(line[j][i])
print(All) 
```

```python
file = open('values.dat','r')
line = []
for i in file:
    line.append(i.split(' '))
All = []

for i in range(len(line[0])-1):
    sum = 0
    temp = [row[i+1] for row in line]
    for j in range(len(temp)-1):
        sum += float(temp[j+1])
    All.append([line[0][i+1],sum])

Givers = []
Takers = []
Zeroes = []
for i in range(len(line[0])-1):
    sum = 0
    temp = [row[i+1] for row in line]
    for j in range(len(temp)-1):
        sum += float(temp[j+1])
    if sum > 0:
        Takers.append([line[0][i+1],sum])
    if sum<0:
        Givers.append([line[0][i+1],sum])
    else:
        Zeroes.append([line[0][i+1],sum])
        
Givers = sorted(Givers,key=itemgetter(1),reverse=True)
Takers = sorted(Takers,key=itemgetter(1))
```
