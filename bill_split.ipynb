```python
file = open('values.dat','r')
line = []
for i in file:
    line.append(i.split(' ')[:-1])
All = []
for i in range(len(line[0])-1):
    sum = 0
    temp = [row[i+1] for row in line]
    for j in range(len(temp)-1):
        sum += float(temp[j+1])
    All.append([line[0][i+1],sum])

Givers = []
Takers = []
Zeroes = []
for i in range(len(line[0])-1):
    sum = 0
    temp = [row[i+1] for row in line]
    for j in range(len(temp)-1):
        sum += float(temp[j+1])
    if sum > 0:
        Takers.append([line[0][i+1],sum])
    if sum<0:
        Givers.append([line[0][i+1],sum])
    else:
        Zeroes.append([line[0][i+1],sum])
        
Givers = sorted(Givers,key=itemgetter(1),reverse=True)
Takers = sorted(Takers,key=itemgetter(1))
```
